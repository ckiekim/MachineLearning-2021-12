- 시리즈 메소드

```python
np.random.seed(2021)
dice = pd.Series(np.random.randint(1,7,100))
```

```python
# element의 갯수
dice.count()
```

```python
# 고유한 값
dice.unique()
```

```python
# 고유한 값의 갯수
dice.nunique()
```

```python
# 고유한 값(카테고리)별 갯수 ***********
dice.value_counts()
```

```python
# 합, 평균, 최대, 최소, 표준편차
dice.sum(), dice.mean(), dice.max(), dice.min(), dice.std()
```

```python
# 요약 통계
dice.describe()
```
