```python
# 데이터 전체중에 결측치가 있는지 확인
titanic.isna().sum().sum()
```

```python
# 어느 열에서 결측치가 있는지 확인
titanic.isnull().sum()
```

```python
# 대체하는 방법 - 카테고리형일 경우에는 최빈값
titanic.embark_town.value_counts()
```

```python
titanic.embark_town = titanic.embark_town.fillna('Southampton')
titanic.embark_town.isna().sum()
```

```python
# 숫자 - 평균, 중앙값
# age를 평균으로 대체
titanic.age.fillna(titanic.age.mean(), inplace=True)
titanic.age.isna().sum()
```

```python
# 너무 많은 결측치가 있는 열 - 삭제
titanic.drop(columns=['deck'], inplace=True)
titanic.isna().sum().sum()
```

```python
# dropna() - 결측치가 포함된 행을 삭제
titanic.dropna(inplace=True)
titanic.shape
```

- apply 메소드

```python
df3 = pd.DataFrame({
    'A': [1, 3, 4, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 5, 2, 4, 4]
})
df3
```

```python
def func1(x):
    return x.max() - x.min()
```

```python
df3.apply(func1)
```

```python
df3.apply(lambda x: x.max() - x.min())
```

```python
# 행단위로 하고자 할 경우에는 axis=1 option을 주어야 함
df3.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
# 각 행별로 최대값에서 최소값을 뺀 결과를 '편차'라는 이름의 열로 만든다
df
```

```python
df['편차'] = df.apply(lambda x: x.max() - x.min(), axis=1)
df
```

```python
# 평균이 70점 이상이면 PASS, 아니면 FAIL 인 '결과' 열을 추가할 것
df['결과'] = df.평균.apply(lambda x: 'PASS' if x >= 70 else 'FAIL')
df
```

```python
df['결과2'] = df.apply(lambda x: '합격' if x.평균 >= 70 else '불합격', axis=1)
df
```
