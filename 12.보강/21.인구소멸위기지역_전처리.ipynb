# 인구소멸 위기지역 - 전처리 (Part 1)

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/시군구_성_연령_5세_별_인구2_2020.csv', encoding='euc-kr')
print(df.shape)
df.head(260).tail(5)
```

```python
# 기초자치단체(시군구)에 행정구가 있으면, 행정구 값으로 변경
df['시군구'][df.행정구.notnull()] = df.행정구[df.행정구.notnull()]
df.head(260).tail()
```

```python
# 소계 데이터 지우기
df = df[df['시군구'] != '소계']
print(df.shape)
df.head()
```

```python
# 20~39세, 65세 이상 컬럼 만들기
df['20-39세'] = df['20 - 24세'] + df['25 - 29세'] + df['30 - 34세'] + df['35 - 39세']
df['65세이상'] = df['65 - 69세'] + df['70 - 74세'] + df['75 - 79세'] + df['80세 이상']
df = df[['광역시도','시군구','구분','인구수','20-39세','65세이상']]
df.head()
```

```python
# pivot_table을 만들어서 시군구 단위로 그룹핑
pop = pd.pivot_table(df,
                     index=['광역시도','시군구'],
                     columns=['구분'],
                     values=['인구수','20-39세','65세이상'])
print(pop.shape)
pop.head()
```

```python
# 인구 소멸비율 계산 
pop['소멸비율'] = pop['20-39세','여자'] / pop['65세이상','계']
pop.head(3)
```

```python
pop['소멸위기지역'] = pop.소멸비율 < 0.5
pop['소멸위기고위험지역'] = pop.소멸비율 < 0.2
pop.head(3)
```

```python
# 인구소멸 위기 지역
crisis_region = pop[pop.소멸위기지역].index.get_level_values(1)
print(len(crisis_region))
print(crisis_region)
```

```python
# 인구소멸 위기 고위험지역
high_crisis_region = pop[pop.소멸위기고위험지역].index.get_level_values(1)
print(len(high_crisis_region))
print(high_crisis_region)
```

- 컬럼명 변경(이중 인덱스 --> 단일 인덱스)

```python
pop.columns.get_level_values(0)
```

```python
for i in range(len(pop.columns.get_level_values(0))):
    print(pop.columns.get_level_values(0)[i] + pop.columns.get_level_values(1)[i])
```

```python
new_columns = []
for i in range(len(pop.columns.get_level_values(0))):
    new_columns.append(pop.columns.get_level_values(0)[i] + pop.columns.get_level_values(1)[i])
pop.columns = new_columns
pop.reset_index(inplace=True)
pop['시군구'] = pop['시군구'].str.strip()
pop.head()
```

```python
pop.to_csv('data/시군구_전처리완료.csv', index=False, encoding='euc-kr')
```
